```python
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(3, 4))
spacing = 0.1
#fig.subplots_adjust(bottom=spacing)
# set width of bars
barWidth = 0.2
 
# set heights of bars
bars1 = [0.004187884274870157, 0.0044165113940835]
bars2 = [0.004359671380370855, 0.0042499215342104435]
 
# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
#r3 = [x + barWidth for x in r2]
 
# Make the plot
plt.bar(r1, bars1, color='mediumvioletred', width=barWidth, edgecolor='white', label='LSTM')
plt.bar(r2, bars2, color='forestgreen', width=barWidth, edgecolor='white', label='GRU')
 
# Add xticks on the middle of the group bars
plt.ylabel("Mean Squred Error (MSE)", fontweight="bold")
#plt.xlabel('group', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(bars1))], ['All\nFeatures', 'Reduced\nFeatures'])
 
# Create legend & Show graphic
plt.legend(loc='upper center', bbox_to_anchor=(0.5,-0.20), fancybox=True, shadow=True, ncol=2)
fig.tight_layout()
plt.savefig("images/mse.png",dpi=6000)
plt.show()
```

```python
# Mean Absolute Error
fig = plt.figure(figsize=(3, 4))
spacing = 0.1
fig.subplots_adjust(bottom=spacing)
# set width of bars
barWidth = 0.2
 
# set heights of bars
bars1 = [ 0.0433, 0.0441]
bars2 = [0.0443, 0.0433]
 
# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
#r3 = [x + barWidth for x in r2]
 
# Make the plot
plt.bar(r1, bars1, color='blueviolet', width=barWidth,  label='LSTM')
plt.bar(r2, bars2, color='mediumseagreen', width=barWidth,  label='GRU')
 
# Add xticks on the middle of the group bars
plt.ylabel("Mean Absolute Error (MAE)", fontweight="bold")
#plt.xlabel('group', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(bars1))], ['All\nFeatures', 'Reduced\nFeatures'])
 
# Create legend & Show graphic
plt.legend(loc='upper center', bbox_to_anchor=(0.50,-0.20), fancybox=True, shadow=True, ncol=2)
fig.tight_layout()
plt.savefig("images/mae.png",dpi=6000)
plt.show()
```

```python
# Mean Absolute Error
fig = plt.figure(figsize=(3, 4))
spacing = 0.1
#fig.subplots_adjust(bottom=spacing)
# set width of bars
barWidth = 0.2
 
# set heights of bars
LSTM = [ 89475.8203,  81154.5469]
GRU = [12821.5537, 31295.8164]
 
# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
#r3 = [x + barWidth for x in r2]
 
# Make the plot
plt.bar(r1, LSTM, color='crimson', width=barWidth, label='LSTM')
plt.bar(r2, GRU, color='seagreen', width=barWidth, label='GRU')
 
# Add xticks on the middle of the group bars
plt.ylabel("Mean Absolute Percentage Error\n(MAPE)", fontweight="bold")
#plt.xlabel('group', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(bars1))], ['All\nFeatures', 'Reduced\nFeatures'])
 
# Create legend & Show graphic
plt.legend(loc='upper center', bbox_to_anchor=(0.50,-0.20), fancybox=True, shadow=True, ncol=2)
fig.tight_layout()
plt.savefig("images/mape.png",dpi=6000)
plt.show()
```

```python
# Loss: Mean squred error

x = np.arange(2)
lstm = [0.004187884274870157,0.0044165113940835]
gru = [0.004359671380370855, 0.0042499215342104435]
width = 0.1
plt.bar(x, lstm, width, color='cyan')
plt.bar(x+0.1, gru, width, color='orange')
plt.xticks(x, ['All Features', 'Redeuced Features'])
plt.xlabel("# of Features")
plt.ylabel("Mean Squared Error(MSE)")
plt.legend(["LSTM", "GRU"])
plt.show()
```
